```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "sales_data.csv"
df = pd.read_csv(file)
```

```python
df['FIRST_ORDER_DATE'] = pd.to_datetime(df['FIRST_ORDER_DATE'],format='%d-%b-%y')
df['LAST_ORDER_DATE'] = pd.to_datetime(df['LATEST_ORDER_DATE'],format='%d-%b-%y')
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df['AVGDAYSBETWEENORDERS'],bins=30,kde=True)
plt.title('Distribution of Average Days Between Orders')
plt.xlabel('Average Days Between Orders')
plt.ylabel('Customer Count')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.histplot(df['DAYSSINCELASTORDER'],bins=30,kde=True)
plt.title('Days since last order')
plt.xlabel('Days since last order')
plt.ylabel('Customer Count')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x=df['TOTAL_ORDERS'],y=df['REVENUE'])
plt.title('Customer Segmentations: Orders vs Revenue')
plt.xlabel("Total Orders")
plt.ylabel('Total Revenue')
plt.show()
```

```python
time_slots = ['TIME_0000_0600_ORDERS', 'TIME_0601_1200_ORDERS', 'TIME_1200_1800_ORDERS', 'TIME_1801_2359_ORDERS']
df_time_orders = df[time_slots].sum()
```

```python
plt.figure(figsize=(10, 4))
sns.barplot(x=df_time_orders.index, y=df_time_orders.values)
plt.title('Total Orders by Time Slot')
plt.xlabel('Time Slot')
plt.ylabel('Total Orders')
# plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x=df['TOTAL_ORDERS'], y=df['CARRIAGE_REVENUE'])
plt.title('Shipping Revenue vs Total Orders')
plt.xlabel('Total Orders')
plt.ylabel('Shipping Revenue')
plt.show()
```
