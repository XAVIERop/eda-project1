#### What is EDA?
##### EDA (Exploratory Data Analysis) is the process of analyzing and summarizing a dataset using statistical and visualization techniques to uncover patterns, relationships, and anomalies before applying machine learning models. It helps in understanding data distributions, detecting missing values, and identifying outliers for better decision-making.

#### Visualization
##### Visualization is the presentation of data in graphicall or visual form to understand the data mroe clearly. Visualization is easy to understand the data.
1. Easily understand the features of data
2. Easily analyze the data and summarize it
3. Help to get meaningful insights from data
4. Help to find the trend or pattern of the data

### Steps involved in EDA

#### 1. Data Sourcing : 
1. Process of gathering data from multiple sources as external or internal data collection
2. There are two major kind of data acc to source :
   1. Public Data :  Data which is easy to access without taking any permission from the agenices. 
   2. Private Data : Data which is not available online and we have to permission from org to get access


#### 2. Data Cleaning : 
1. Get rid of any info that doesnt need to be there
2. Process of cleaning the data for further data analysis and builiding ML model.
3. Benefit : all irrelevant data is gone and we get quality data
4. Helps in improving accuracy of ML model
5. Steps :
   1. Handle Missing Values
   2. Feature Scaling : Standardization, Normalization of the data
   3. Outlier Treatment
   4. Handle Invalid values


##### <b>A. Handle Missing Values</b>
1. <b>Delete rows/cols : </b>  Rows can be deleted if it has insignificant number of missing value. Cols can be deleted if it has more than 75% of missing value.
2. <b>Replacing with mean/median/mode :</b> Used on independent variable when it has numerical variables. Mode is used for categorical data.
3. <b>Algorithm Imputation :</b> Some ML algos support to handle missing values like KNN, Naive Bayes, Random Forest.
4. <b>Predicting the missing values :</b> Prediction model is one of the advanced method to handle missing values. In this method dataset with no missing value become training set and dataset with missing value become the test set and the missing values is treated as target variable.
