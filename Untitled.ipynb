```python
# outliers
numeric_cols = df.select_dtypes(include=['number'])
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
outliers = ((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).sum()
print(outliers[outliers > 0])
```

```python
# Select a specific customer (e.g., CustomerID 22)
customer_id = 22
customer_data = df[df['CustomerID'] == customer_id]

# Plot purchase frequency over time
customer_data['FIRST_ORDER_DATE'] = pd.to_datetime(customer_data['FIRST_ORDER_DATE'], format="%d-%b-%y")
customer_data['LATEST_ORDER_DATE'] = pd.to_datetime(customer_data['LATEST_ORDER_DATE'], format="%d-%b-%y")

plt.figure(figsize=(10, 5))
plt.bar(['First Order', 'Last Order'], 
        [customer_data['FIRST_ORDER_DATE'].values[0], customer_data['LATEST_ORDER_DATE'].values[0]], 
        color=['blue', 'red'])
plt.title(f'Customer {customer_id} Order Timeline')
plt.ylabel('Date')
plt.show()

# Plot order frequency by day of the week
order_days = ['MONDAY_ORDERS', 'TUESDAY_ORDERS', 'WEDNESDAY_ORDERS', 
              'THURSDAY_ORDERS', 'FRIDAY_ORDERS', 'SATURDAY_ORDERS', 'SUNDAY_ORDERS']

order_counts = customer_data[order_days].values.flatten()

plt.figure(figsize=(10, 5))
plt.bar(order_days, order_counts, color='green')
plt.title(f'Customer {customer_id} - Orders by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()

# Plot revenue distribution by time slots
time_slots = ['TIME_0000_0600_REVENUE', 'TIME_0601_1200_REVENUE', 
              'TIME_1200_1800_REVENUE', 'TIME_1801_2359_REVENUE']

revenue_values = customer_data[time_slots].values.flatten()

plt.figure(figsize=(10, 5))
plt.bar(time_slots, revenue_values, color='orange')
plt.title(f'Customer {customer_id} - Revenue by Time Slot')
plt.xlabel('Time Slot')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.show()
```
